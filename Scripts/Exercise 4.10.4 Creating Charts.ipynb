```python
# Importing libraries 

import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
import seaborn as sns
import scipy
```

```python
# Importing data

path = r'C:\Users\katro\Instacart Basket Analysis'
```

```python
df_ords = pd.read_pickle(os.path.join(path, '02 Data', 'Prepared Data', 'data_combined_3.pkl'))
```

```python
# Creating chart for busiest day of the week

bar = df_ords['order_day_of_week'].value_counts().sort_index().plot.bar()
```

```python
# Exporting bar chart

bar.figure.savefig(os.path.join(path, '04 Analysis', 'Visualizations', 'bar_dow.png'))
```

```python
# Creating bar chart to show busiest time of the day for orders

bar_2 = df_ords['order_hour_of_day'].value_counts().sort_index().plot.bar()
```

```python
# Exporting bar chart

bar_2.figure.savefig(os.path.join(path, '04 Analysis', 'Visualizations', 'bar_hours.png'))
```

```python
# Importing department data

df_deps = pd.read_csv(os.path.join(path, '02 Data', 'Prepared Data', 'departments_wrangled.csv'))
```

```python
# Checking columns

df_deps.columns
```

```python
# Merging department data with df_ords

df_merge = df_ords.merge(df_deps, on = 'department_id', how = 'left')
```

```python
# Creating bar chart comparing departments

bar_3 = df_merge['department'].value_counts().plot.barh()
```

```python
# Exporting departments bar chart

bar_3.figure.savefig(os.path.join(path, '04 Analysis', 'Visualizations', 'bar_depts.png'))
```

```python
# Creating bar graph comparing brand loyalty

bar_4 = df_merge['loyalty flag'].value_counts().plot.barh()
```

```python
# Exporting loyalty bar chart

bar_4.figure.savefig(os.path.join(path, '04 Analysis', 'Visualizations', 'bar_loyalty.png'))
```

```python
# Exporting new department dataframe

df_merge.to_pickle(os.path.join(path, '02 Data', 'Prepared Data', 'data_combined_4.pkl'))
```
