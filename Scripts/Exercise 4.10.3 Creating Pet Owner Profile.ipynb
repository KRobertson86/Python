```python
# Importing libraries 

import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
import seaborn as sns
import scipy
```

```python
# Importing data

path = r'C:\Users\katro\Instacart Basket Analysis'
```

```python
df_ords = pd.read_pickle(os.path.join(path, '02 Data', 'Prepared Data', 'data_combined_3.pkl'))
```

```python
# Creating pet owner profile

df_pet_owner = df_ords[df_ords['department_id'] == 8]
```

```python
df_pet_owner.shape
```

```python
# Requesting to just see three columns of pet owners

df_pet_owner[["age", "income", "n_dependants"]]
```

```python
# Creating age and number of dependants graph

line = sns.lineplot(data = df_pet_owner, x='age', y='n_dependants')
```

```python
# Creating graph comparing age with income

splot = sns.scatterplot(x='age', y='income', data=df_pet_owner)
```

```python
# Mean, minimum, and maximum of 'prices' in pet owners profile

df_pet_owner.groupby('department_id').agg({'prices':['mean', 'min', 'max']})
```

```python
# Mean, minimim, and maximum of 'days_since_prior_order' in pet owners profile

df_pet_owner.groupby('department_id').agg({'days_since_prior_order':['mean', 'min', 'max']})
```

```python
# Comparing pet owners across regions

bar = df_pet_owner['region'].value_counts().plot.bar()
```

Most pet owners are in the South and the least pet owners are in the Northeast.

```python
# Exporting chart

bar.figure.savefig(os.path.join(path, '04 Analysis', 'Visualizations', 'region_pets.png'))
```
